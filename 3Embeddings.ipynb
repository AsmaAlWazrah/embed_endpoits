```python
#import pandas as pd
import pyarabic.araby as arab
from transformers import pipeline
```

```python
electrapipeline = pipeline('feature-extraction', model='aubmindlab/araelectra-base-generator')
data = electrapipeline("ضرب")
print(data)
```

```python
arabertpipeline = pipeline('feature-extraction', model='aubmindlab/bert-base-arabertv2')
data = arabertpipeline("مرحبا")
print(data)
```

```python
camelpipeline = pipeline('feature-extraction', model='CAMeL-Lab/bert-base-arabic-camelbert-msa')
data = camelpipeline("ضرب")
print(data)
```

```python
import numpy as np
import re
for i in range(len(Train["wordlemma"])):
  if len(Train.loc[i,'wordlemma'].split())== 1:
    token=Train.loc[i,'wordlemma']
    token = re.sub(r"[^\u0600-\u06ff ]", "", token)
    token=arab.strip_tashkeel(token)
    Train["electra"].loc[i]= electrapipeline(token)[0][1]
    Train["bertseg"].loc[i]= arabertpipeline(token)[0][1]
    Train["bertmsa"].loc[i]= camelpipeline(token)[0][1]
    print(i);

  else:
    # print(i);
    token=Train.loc[i,'wordlemma']
    token = re.sub(r"[^\u0600-\u06ff ]", "", token)
    token=arab.strip_tashkeel(token)
    # print(token)
    emb=electrapipeline( arab.strip_tashkeel(token))[0][1:-1]
    Train["electra"].loc[i]= np.mean(emb, axis=0)

    araemb=arabertpipeline( arab.strip_tashkeel(token))[0][1:-1]
    Train["bertseg"].loc[i]= np.mean(araemb, axis=0)

    camemb=camelpipeline( arab.strip_tashkeel(token))[0][1:-1]
    Train["bertmsa"].loc[i]= np.mean(camemb, axis=0)
```
